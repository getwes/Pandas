### Importando o pandas

```python
import pandas as pd
```

### Criando um dataframe a partir de um dicionário

```python
# dataframe = pd.DataFrame()
venda = {'data': ['15/02/2021', '16/02/2021'],
         'valor': [500, 300],
         'produto': ['feijao', 'arroz'],
         'qtde': [50, 70],
        }
#df é a abreviação do dataframe df
vendas_df = pd.DataFrame(venda)
```

### Visualização dos Dados
    - print
    - display

```python
display(vendas_df)
```

### Importando arquivos e bases de dados

```python
vendas_df = pd.read_excel("vendas.xlsx")
display(vendas_df)
```

### Resumos de Visualização de Dados simples e úteis
    - head
    - shape
    - describe

```python
display(vendas_df.head(10)) # posso ver quantas tabelas for necessaria atraves do indice dentro do head
print(vendas_df.shape)# quantidade de linha e coluna de forma muito mais rapida do que o display
display(vendas_df.describe()) #faz a resumo de todos os valores numericos
```

### Pegar 1 coluna (e os pd.Series)

```python
produtos = vendas_df[['Produto', 'ID Loja', 'Valor Unitário']]
display(produtos)
```

### .loc, um método muito importante
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# pegar uma linha
display(vendas_df.loc[1:5]) #ver todos os dados da coluna 1 ate a 5

#pegar linhas que correspondem a uma condição
vendas_df['ID Loja'] == 'Norte Shopping'

display(vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping'])

#posso armazenar os dados tambem
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']

#pegar varias linhas e colunas usando o loc
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ["ID Loja", "Produto", "Quantidade"]]
display(vendas_norteshopping_df)

# pegar 1 valor especifico
print(vendas_df.loc[3, 'Produto'])
```
